```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
a_sales_20_path= r'Resources/athletic_sales_2020.csv'
a_sales_21_path = r'Resources/athletic_sales_2021.csv'

a_sales20_data = pd.DataFrame(pd.read_csv(a_sales_20_path))
a_sales21_data = pd.DataFrame(pd.read_csv(a_sales_21_path))
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_years_df = pd.concat([a_sales20_data, a_sales21_data], axis=0)

combined_years_df
c_yrs_df = combined_years_df.copy()
c_yrs_df
```

```python
# Check if any values are null.
data_set_has_null = c_yrs_df.isnull().sum()
data_set_has_null
```

```python
# Check the data type of each column
c_yrs_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
c_yrs_df['invoice_date'] =  pd.to_datetime(c_yrs_df['invoice_date'], format='%m/%d/%y')
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(c_yrs_df.info())
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
ops_market = ['region', 'state', 'city']
unit_sold_info = 'units_sold'
grouped_ops_market = c_yrs_df.groupby(ops_market, as_index=False).agg(
    Total_Products_Sold=(unit_sold_info, 'sum')
)
grouped_ops_market = grouped_ops_market.rename(columns={'region': 'Region', 'state': 'State', 'city': 'City'})
regional_product_info = grouped_ops_market.sort_values(by='Total_Products_Sold', ascending=False).reset_index()

# Show the top 5 results.
regional_product_info.head()

```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
ops_market = ['region', 'state', 'city']
pivot_unit_sold = c_yrs_df.pivot_table(
    index=ops_market,
    values=unit_sold_info,
    aggfunc='sum'
).rename(columns={unit_sold_info: 'Total_Products_Sold'}).reset_index()
pivot_ops_market = pivot_unit_sold.rename(columns={'region': 'Region', 'state': 'State', 'city': 'City'})
regional_product_info = pivot_ops_market.sort_values(by='Total_Products_Sold', ascending=False).reset_index()
# Show the top 5 results.
regional_product_info.head()
```

```python
#Region Sold the Most Products
region_names = regional_product_info['Region']
top_region = region_names.head(1).to_string(index=False)
print(f'During operating time period of both 2020 and 2021 {top_region} region had the most product(s) sold')
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales_info = 'total_sales'
total_sales = c_yrs_df.groupby(ops_market, as_index=False).agg(
    Total_Sales=(total_sales_info, 'sum')
)
total_sales.columns
ops_market_sales = total_sales.rename(columns={'region': 'Region', 'state': 'State', 'city': 'City'})
regional_sales_info = ops_market_sales.sort_values(by='Total_Sales', ascending=False).reset_index()

# Show the top 5 results.
regional_sales_info.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
pivot_total_sales = c_yrs_df.pivot_table(
    index=ops_market,
    values=total_sales_info,
    aggfunc='sum'
).rename(columns={'total_sales': 'Total Sales'}).reset_index()
pivot_total_sales = pivot_total_sales.rename(columns={'region': 'Region', 'state': 'State', 'city': 'City'})
regional_sales_info = pivot_total_sales.sort_values(by='Total Sales', ascending=False).reset_index()
# # Show the top 5 results.
regional_sales_info.head()
```

```python
#Region with most Sales
region_names = regional_sales_info['Region']
top_region = region_names.head(1).to_string(index=False)
print(f'During operating period of both 2020 and 2021 {top_region} region had the most sales.')
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
retailer_info = ['retailer', 'state', 'city']
total_retailer_sales = c_yrs_df.groupby(retailer_info, as_index=False).agg(
    Total_Sales=(total_sales_info, 'sum')
)
total_retailer_sales = total_retailer_sales.rename(columns={'retailer':'Retailer','region': 'Region', 'state': 'State', 'city': 'City'})
retailer_sales_info  = total_retailer_sales.sort_values(by='Total_Sales', ascending=False)
retailer_sales_info.head()

# Show the top 5 results.
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_retailer_sales = c_yrs_df.pivot_table(
    index=retailer_info,
    values=total_sales_info,
    aggfunc='sum'
).rename(columns={'total_sales': 'Total Sales'}).reset_index()
pivot_retailer_sales = pivot_retailer_sales.rename(columns={'retailer':'Retailer', 'region': 'Region', 'state': 'State', 'city': 'City'})
retailer_sales_info = pivot_retailer_sales.sort_values(by='Total Sales', ascending=False).reset_index()
# Show the top 5 results.
retailer_sales_info.head()
```

```python
#Retailer with most Sales
retailer_names = retailer_sales_info['Retailer']
top_retailer = retailer_names.head(1).to_string(index=False)
print(f'During operating period of both 2020 and 2021 {top_retailer} retailer had the most sales.')
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
target_word = "Women's Athletic Footwear"
women_athletic_footware = c_yrs_df[c_yrs_df['product']==target_word]
women_athletic_footware.index
women_athletic_footware
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_athletic_footware_sales = women_athletic_footware.groupby(retailer_info, as_index=False).agg(
    Womens_Footwear_Units_Sold=(unit_sold_info, 'sum')
)
women_athletic_footware_sales = women_athletic_footware_sales.rename(columns={'retailer':'Retailer','region': 'Region', 'state': 'State', 'city': 'City'})
women_athletic_footware_sales  = women_athletic_footware_sales.sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
women_athletic_footware_sales.head()

# Show the top 5 results.
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_athletic_footware_sales = women_athletic_footware.pivot_table(
    index=retailer_info,
    values=unit_sold_info,
    aggfunc='sum'
).rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}).reset_index()
women_athletic_footware_sales = women_athletic_footware_sales.rename(columns={'retailer':'Retailer', 'region': 'Region', 'state': 'State', 'city': 'City'})
women_athletic_footware_sales = women_athletic_footware_sales.sort_values(by='Womens_Footwear_Units_Sold', ascending=False).reset_index()
# Show the top 5 results.
women_athletic_footware_sales.head()
```

```python
#Women's Athletic Footwear Sold
retailer_names = women_athletic_footware_sales['Retailer']
top_retailer = retailer_names.head(1).to_string(index=False)
print(f'During operating period of both 2020 and 2021 {top_retailer} retailer had the most women athletic footware sales.')
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"
w_ftwr_sales_invoice_date = women_athletic_footware.pivot_table(
    index='invoice_date',
    values=total_sales_info,
    aggfunc='sum'
).rename(columns={'total_sales': 'Total Sales'}).reset_index()
w_ftwr_sales_info = w_ftwr_sales_invoice_date.set_index('invoice_date')
# Show the table.
w_ftwr_sales_info
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales = w_ftwr_sales_info.resample('D').sum()
sorted_daily_sales = daily_sales.sort_values(by='Total Sales', ascending=False)
sorted_daily_sales.head(10)
```

```python
# Days with Women's Athletic Footwear Sold
invoice_dates = daily_sales.index
top_days = invoice_dates[0]
print(f'During operating period of both 2020 and 2021 {top_days} was the day with the most women athletic footware sales.')
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales = w_ftwr_sales_info.resample('W').sum()
weekly_sales = weekly_sales.sort_values(by='Total Sales', ascending=False)
weekly_sales.head(10)
```

```python
#Weeks with Women's Athletic Footwear Sold
invoice_dates = weekly_sales.index
top_days = invoice_dates[0]
print(f'During operating period of both 2020 and 2021 {top_days} was the week with the most women athletic footware sales.')
```
