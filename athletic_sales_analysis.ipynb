```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
a_sales_20_path= r'Resources/athletic_sales_2020.csv'
a_sales_21_path = r'Resources/athletic_sales_2021.csv'

a_sales20_data = pd.DataFrame(pd.read_csv(a_sales_20_path))
a_sales21_data = pd.DataFrame(pd.read_csv(a_sales_21_path))
```

```python
# Display the 2021 sales DataFrame
a_sales21_data.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
a_sales20_data.info()
```

```python
# Check the 2021 sales data types.
a_sales21_data.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_years_df = pd.merge(a_sales20_data, a_sales21_data, on='retailer_id', how='inner', suffixes=('_2020', '_2021')).reset_index()

combined_years_df
c_yrs_df = combined_years_df.copy()
c_yrs_df.head()
```

```python
# Check if any values are null.
data_set_has_null = c_yrs_df.isnull().sum()
data_set_has_null
```

```python
# Check the data type of each column
c_yrs_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
invoice_dates = ['invoice_date_2020', 'invoice_date_2021']
date = c_yrs_df[invoice_dates]
c_yrs_df['invoice_date_2020'] =  pd.to_datetime(c_yrs_df['invoice_date_2020'], format='%m/%d/%y')
c_yrs_df['invoice_date_2021'] =  pd.to_datetime(c_yrs_df['invoice_date_2021'], format='%m/%d/%y')
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(c_yrs_df.info())
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

target_market_2020 = ['region_2020', 'state_2020', 'city_2020']
product_sales_info_2020 = 'units_sold_2020'
prod_sold_2020 = c_yrs_df.groupby(target_market_2020, as_index=False).agg(
    Total_Products_Sold=(product_sales_info_2020, 'sum')
)
target_market_2021 = ['region_2021', 'state_2021', 'city_2021']
product_sales_info_2021 = 'units_sold_2021'
prod_sold_2021 = c_yrs_df.groupby(target_market_2021, as_index=False).agg(
    Total_Products_Sold=(product_sales_info_2021, 'sum')
)
# Show the top 5 results.

prod_sold_2020 = prod_sold_2020.rename(columns={'region_2020': 'Region', 'state_2020': 'State', 'city_2020': 'City'})
prod_sold_2021 = prod_sold_2021.rename(columns={'region_2021': 'Region', 'state_2021': 'State', 'city_2021': 'City'})
consolidate_sales_info =  pd.concat([prod_sold_2020,prod_sold_2021], ignore_index=True)
consolidate_sales_info = consolidate_sales_info.sort_values(by='Total_Products_Sold', ascending=False).reset_index()
consolidate_sales_info.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
prod_sold_pivot_2020 = c_yrs_df.pivot_table(
    index=['region_2020', 'state_2020', 'city_2020'],
    values='units_sold_2020',
    aggfunc='sum'
).rename(columns={'units_sold_2020': 'Total_Products_Sold'}).reset_index()

prod_sold_pivot_2021 = c_yrs_df.pivot_table(
    index=['region_2021', 'state_2021', 'city_2021'],
    values='units_sold_2021',
    aggfunc='sum'
).rename(columns={'units_sold_2021': 'Total_Products_Sold'}).reset_index()

prod_sold_pivot_2020 = prod_sold_pivot_2020.rename(columns={'region_2020': 'Region', 'state_2020': 'State', 'city_2020': 'City'})
prod_sold_pivot_2021 = prod_sold_pivot_2021.rename(columns={'region_2021': 'Region', 'state_2021': 'State', 'city_2021': 'City'})
consolidate_pivot_sales_info =  pd.concat([prod_sold_pivot_2020,prod_sold_pivot_2021], ignore_index=True)

consolidate_pivot_sales_info = consolidate_pivot_sales_info.sort_values(by='Total_Products_Sold', ascending=False).reset_index()
# Show the top 5 results.
consolidate_pivot_sales_info.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
target_market_2020 = ['region_2020', 'state_2020', 'city_2020']
total_sales_info_2020 = 'total_sales_2020'
total_sales_2020 = c_yrs_df.groupby(target_market_2020, as_index=False).agg(
    Total_Sales=(total_sales_info_2020, 'sum')
)
target_market_2021 = ['region_2021', 'state_2021', 'city_2021']
product_sales_info_2021 = 'total_sales_2021'
total_sales_2021 = c_yrs_df.groupby(target_market_2021, as_index=False).agg(
    Total_Sales=(product_sales_info_2021, 'sum')
)
# Show the top 5 results.

total_sales_2020 = total_sales_2020.rename(columns={'region_2020': 'Region', 'state_2020': 'State', 'city_2020': 'City'})
total_sales_2021 = total_sales_2021.rename(columns={'region_2021': 'Region', 'state_2021': 'State', 'city_2021': 'City'})
consolidate_sales_info =  pd.concat([total_sales_2020,total_sales_2021], ignore_index=True)
consolidate_sales_info = consolidate_sales_info.sort_values(by='Total_Sales', ascending=False).reset_index()
consolidate_sales_info.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"
prod_sales_pivot_2020 = c_yrs_df.pivot_table(
    index=['region_2020', 'state_2020', 'city_2020'],
    values='total_sales_2020',
    aggfunc='sum'
).rename(columns={'total_sales_2020': 'Total Sales'}).reset_index()

prod_sales_pivot_2021 = c_yrs_df.pivot_table(
    index=['region_2021', 'state_2021', 'city_2021'],
    values='total_sales_2021',
    aggfunc='sum'
).rename(columns={'total_sales_2021': 'Total Sales'}).reset_index()

prod_sales_pivot_2020 = prod_sales_pivot_2020.rename(columns={'region_2020': 'Region', 'state_2020': 'State', 'city_2020': 'City'})
prod_sales_pivot_2021 = prod_sales_pivot_2021.rename(columns={'region_2021': 'Region', 'state_2021': 'State', 'city_2021': 'City'})
consolidate_pivot_sales_info =  pd.concat([prod_sales_pivot_2020,prod_sales_pivot_2021], ignore_index=True)
consolidate_pivot_sales_info = consolidate_pivot_sales_info.sort_values(by='Total Sales', ascending=False).reset_index()
# Show the top 5 results.
consolidate_pivot_sales_info.head()
```
